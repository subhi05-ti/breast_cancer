```python
# importing the dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Separating the features and target

```python
x = breast_cancer_data.drop(columns='label',axis=1)
y = breast_cancer_data['label']
```

```python
print(x)
```

```python
print(y)
```

Train_Test_Split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.fit_transform(x_test)
```

```python
print(x_train_std)
```

**Building the Neural Network**
libraries used to built neural network:-

tensorflow,keras,pytorch

```python
# importing the tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting uo the layer of nneural network
model = keras.Sequential([
                           keras.layers.Flatten(input_shape=(30,)),  # input layer (falttenconvert the multi-d array in a single dimension)
                           keras.layers.Dense(20,activation='relu'),    # one hidden layer
                           #keras.layers.Dense(30,activation='sigmoid'),   # for adding another hidden llayer
                           keras.layers.Dense(2,activation='sigmoid')    # one output layer
])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])
```

```python
# training the neural network
history = model.fit(x_train_std,y_train,validation_split=0.1,epochs=15)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation data'],loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model_loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data','validation data'],loc = 'upper right')
```

Accuracy of the model on test data

```python
loss,accuracy = model.evaluate(x_test_std,y_test)
print(accuracy*100)
```
